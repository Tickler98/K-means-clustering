Assignment 3: K-means Clustering

```python
#Importing the necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score
```

```python
#Loading the dataset (same as assignment 2)
customers = pd.read_csv("/content/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
# Printing the information of the dataset
print(customers.info())
```

```python
#Selecting the features
X = customers[['MonthlyCharges', 'tenure']].copy()
```

```python
# Scater plot the selected features
sns.scatterplot(data=customers, x='MonthlyCharges', y='tenure')
plt.title('Monthly Charges vs Tenure')
plt.show()
```

```python
# Features standardising
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Using three different methods (Elbow method, Silhouette Score, and Calinski-Harabasz Score) to get the K(cluster)

```python
# Elbow Method
wcss = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=1234)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(2, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.show()
```

```python
# Silhouette Score method
silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=1234)
    kmeans.fit(X_scaled)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

plt.plot(range(2, 11), silhouette_scores)
plt.title('Silhouette Score')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
# Calinski-Harabasz Score method
calinski_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=1234)
    kmeans.fit(X_scaled)
    calinski_scores.append(calinski_harabasz_score(X_scaled, kmeans.labels_))

plt.plot(range(2, 11), calinski_scores)
plt.title('Calinski-Harabasz Score')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Calinski-Harabasz Score')
plt.show()
```

```python
# Creating clusters with K=4 (based on all the 3 methods)
kmeans = KMeans(n_clusters=4, random_state=1234)
kmeans.fit(X_scaled)
customers['Cluster'] = kmeans.labels_
```

```python
# Visualising the clusters
sns.scatterplot(data=customers, x='MonthlyCharges', y='tenure', hue='Cluster', palette='colorblind')
plt.title('Clusters of Customers (Monthly Charges vs Tenure)')
plt.show()
```
